```python
def kg():
    weight = int(input("Enter your weight in kg: "))
    height = int(input("Enter your height in cm: "))
    cnv_mtr = height / 100
    sqr_height = cnv_mtr * cnv_mtr
    bmi = weight / sqr_height

    print("\nYour BMI is:", round(bmi, 2))
    print("BMI Classification:")

    if bmi <= 16:
        print("Severe Thinness")
    elif 16 < bmi <= 17:
        print("Moderate Thinness")
    elif 17 < bmi <= 18.5:
        print("Mild Thinness")
    elif 18.5 < bmi <= 25:
        print("Normal")
    elif 25 < bmi <= 30:
        print("Overweight")
    else:
        print("Obese")

        
def pound():
    weight = int(input("Enter your weight in pounds: "))
    height = int(input("Enter your height in cm: "))
    # Convert pounds to kg
    weight_kg = weight * 0.453592
    cnv_mtr = height / 100
    sqr_height = cnv_mtr * cnv_mtr
    bmi = weight_kg / sqr_height

    print("\nYour BMI is:", round(bmi, 2))
    print("BMI Classification:")

    if bmi <= 16:
        print("Severe Thinness")
    elif 16 < bmi <= 17:
        print("Moderate Thinness")
    elif 17 < bmi <= 18.5:
        print("Mild Thinness")
    elif 18.5 < bmi <= 25:
        print("Normal")
    elif 25 < bmi <= 30:
        print("Overweight")
    else:
        print("Obese")

# Main Program
catg = input("Press K for Kg / Press P for Pound: ").upper()

if catg == "K":
    kg()
elif catg=="P":
    pound()
else:
    print("Sorry! Please choose a valid option.")
```

```python
import tkinter as tk
from tkinter import messagebox

# Function for BMI calculation in kg
def calculate_bmi_kg():
    try:
        weight = float(entry_weight.get())
        height = float(entry_height.get())
        cnv_mtr = height / 100
        sqr_height = cnv_mtr * cnv_mtr
        bmi = weight / sqr_height
        display_result(bmi)
    except ValueError:
        messagebox.showerror("Invalid input", "Please enter valid numbers for weight and height.")

# Function for BMI calculation in pounds
def calculate_bmi_pounds():
    try:
        weight = float(entry_weight.get())
        height = float(entry_height.get())
        weight_kg = weight * 0.453592  # Convert pounds to kg
        cnv_mtr = height / 100
        sqr_height = cnv_mtr * cnv_mtr
        bmi = weight_kg / sqr_height
        display_result(bmi)
    except ValueError:
        messagebox.showerror("Invalid input", "Please enter valid numbers for weight and height.")

# Function to display BMI result and classification
def display_result(bmi):
    bmi_value.set(f"Your BMI is: {round(bmi, 2)}")

    if bmi <= 16:
        classification.set("Classification: Severe Thinness")
    elif 16 < bmi <= 17:
        classification.set("Classification: Moderate Thinness")
    elif 17 < bmi <= 18.5:
        classification.set("Classification: Mild Thinness")
    elif 18.5 < bmi <= 25:
        classification.set("Classification: Normal")
    elif 25 < bmi <= 30:
        classification.set("Classification: Overweight")
    else:
        classification.set("Classification: Obese")

# Main GUI window
root = tk.Tk()
root.title("BMI Calculator")

# Input fields
tk.Label(root, text="Enter your weight:").grid(row=0, column=0, padx=10, pady=5)
entry_weight = tk.Entry(root)
entry_weight.grid(row=0, column=1, padx=10, pady=5)

tk.Label(root, text="Enter your height (cm):").grid(row=1, column=0, padx=10, pady=5)
entry_height = tk.Entry(root)
entry_height.grid(row=1, column=1, padx=10, pady=5)

# Variable to store the results
bmi_value = tk.StringVar()
classification = tk.StringVar()

# Buttons for KG and Pound options
tk.Button(root, text="Calculate BMI (Kg)", command=calculate_bmi_kg).grid(row=2, column=0, padx=10, pady=10)
tk.Button(root, text="Calculate BMI (Pound)", command=calculate_bmi_pounds).grid(row=2, column=1, padx=10, pady=10)

# Result labels
tk.Label(root, textvariable=bmi_value).grid(row=3, column=0, columnspan=2, padx=10, pady=5)
tk.Label(root, textvariable=classification).grid(row=4, column=0, columnspan=2, padx=10, pady=5)

root.mainloop()
```
